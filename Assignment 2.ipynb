# II Dataset: loan.csv

```python
loan = pd.read_csv('loan.csv')
loan.head(2)
```

#### Q1. Display the mean of any two variables with continous values 

```python
print('Mean of disbursed amount: {:.2f}\nMean of interest: {:.2f}'.format(loan['disbursed_amount'].mean(),loan['interest'].mean()))
```

#### Q2. Display the number of discrete variables 

```python
# loan.info()
print('Discreet Varibles:\n',[i for i in loan.columns if loan[i].dtypes != 'float64'])
print('\nNumber of discreet variables: ',len([i for i in loan.columns if loan[i].dtypes != 'float64']))
```

#### Q3. Display the unique values of two variables with discreete values 

```python
print('Market: {}\nCustomer_ID: {}'.format(loan['market'].unique(),loan['customer_id'].unique()))
```

#### Q4. Display the Month with most of loans issued date 

```python
loan['date_issued_dt']=pd.to_datetime(loan['date_issued'])
x = loan['date_issued_dt'].dt.month.value_counts()
print(x)
for i in range(1,12):
    if x[i]==x.max():
        print('\nMonth with most loans issued date is',i)
```

#### Q5. Display the count of 'Teacher' who are 'Owners' 

```python
count = 0
for t,v in zip(loan['employment'],loan['householder']):
    if t=='Teacher' and v=='RENT':
        count+=1
print(f'Teacher who are Owners: {count}')
```
